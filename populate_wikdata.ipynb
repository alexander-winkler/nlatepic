```python
import requests
import json
import csv
```

```python
# get request to wikidata api
# input is search string, output is a json object

def wiki_request(search_term):
    endpoint = "https://www.wikidata.org/w/api.php"
    p =  {
        'action' : 'query',
        'srsearch' : search_term,
        'list' : 'search',
        'srlimit' : '2',
        'format' : 'json',
    }
    return requests.get(endpoint,params=p).json()
```

```python
# get info from wikidata entry, given the wikidata id

def get_entry(wikidataId):
    endpoint = "https://www.wikidata.org/w/api.php"
    p = {
        "format" : "json",
        "action" : "wbgetentities",
        "ids" : wikidataId
    }
    return requests.get(endpoint,params=p).json()
```
